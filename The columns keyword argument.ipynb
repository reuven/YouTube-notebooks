# The `columns` keyword argument

```python
import pandas as pd
```

```python
df = pd.read_csv('taxi.csv')
```

```python
df.head()
```

```python
# If I want to remove one of the columns (VendorID), how can I do that?
# df.drop!

df.drop('VendorID')
```

```python
df.drop('VendorID', axis=1)
```

```python
df.drop('VendorID', axis='columns')
```

```python
df.drop(['VendorID', 'passenger_count'], axis='columns')
```

```python
df.drop(columns=['VendorID', 'passenger_count'])
```

```python
# If I want to rename one or more columns, I can do that with df.rename

df.rename(columns={'VendorID':'vendor_id', 'trip_distance':'trip_distance_in_miles'})
```

```python
df.rename({'VendorID':'vendor_id', 'trip_distance':'trip_distance_in_miles'}, axis='columns')
```
