# Three reasons to make a datetime column your index

1. Retrieve with a subset of the datetime
2. You can get time slices
3. You can resample

```python
import pandas as pd

df = pd.read_csv('taxi.csv',
                 parse_dates=['tpep_pickup_datetime'],
                 index_col='tpep_pickup_datetime')
df
```

```python
df.index
```

```python
# subset of the datetime

df.loc['2015-06-01']
```

```python
# 2 get a slice based on the dates and times

df.sort_index().loc['2015-06-01 00:13:02':'2015-06-01 00:13:04']
```

```python
# 3. resampling

# I want to know the mean trip_distance

df['trip_distance'].mean()
```

```python
# I want to know the mean trip_distance for each 24-hour period in our data set

df.resample('1D')['trip_distance'].mean()
```

```python
df.resample('3D')['trip_distance'].mean()
```

```python
df.resample('8h')['trip_distance'].mean()
```
