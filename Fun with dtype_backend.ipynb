```python
# the int64 dtype in NumPy is not "nullable"
# I can use Pandas "extension types."

pd.NA
```

```python
df = pd.read_csv('taxi.csv', dtype_backend='numpy_nullable')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.loc[0, 'passenger_count'] = pd.NA
```

```python
df.head()
```

```python
# the extension types are nullable -- we can use pd.NA along with the integers
```
