# Using `melt` in Pandas

```python
import pandas as pd
```

```python
filename = 'sat-scores.csv'
```

```python
df = pd.read_csv(filename)
df
```

```python
df = pd.read_csv(filename, usecols=['Year', 'State.Code', 'Total.Math', 'Total.Test-takers', 'Total.Verbal'])
df
```

```python
df.melt(id_vars='Year',
        value_vars=['Total.Math', 'Total.Verbal'])
```

```python
df.melt(id_vars='Year',
        value_vars=['Total.Math', 'Total.Verbal'],
       var_name='subject',
       value_name='score')
```

```python
df.melt(id_vars=['Year', 'State.Code'],
        value_vars=['Total.Math', 'Total.Verbal'],
       var_name='subject',
       value_name='score')
```

```python
(
    df
    .melt(id_vars=['Year', 'State.Code'],
        value_vars=['Total.Math', 'Total.Verbal'],
       var_name='subject',
       value_name='score')
    .loc[lambda df_: df_['State.Code'] == 'AL']
    .sort_values(['Year', 'State.Code'])
)
```

```python
(
    df
    .melt(id_vars=['Year', 'State.Code'],
        value_vars=['Total.Math', 'Total.Verbal'],
       var_name='subject',
       value_name='score')
    .loc[lambda df_: df_['State.Code'] == 'AL']
    .sort_values(['Year', 'State.Code'])
).pivot(index=['Year', 'State.Code'],
        columns='subject',
        values='score')
```
