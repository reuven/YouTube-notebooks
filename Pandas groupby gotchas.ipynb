```python
import pandas as pd

df = pd.read_csv('taxi.csv')
```

```python
df.head()
```

# 1. Forgetting to aggregate!


```python
# for every distinct value of passenger_count
# calculate the mean trip_distance

df.groupby('passenger_count')['trip_distance'].mean()
```

```python
df.groupby('passenger_count')['trip_distance']
```

```python
df.groupby('passenger_count').mean()
```

```python
df.groupby('passenger_count').count()
```

# 2. Ignoring sort=False

```python
df.groupby('passenger_count')['trip_distance'].mean()
```

```python
df.groupby('passenger_count', sort=False)['trip_distance'].mean()
```

# 3. Misusing multiple aggregation methods

```python
# what if I want both mean and std?

df.groupby('passenger_count')['trip_distance'].agg(['mean', 'std'])
```

```python
df.groupby('passenger_count').agg({'trip_distance':['mean', 'std'],
                                   'total_amount':['mean', 'std', 'min', 'max']})
```

# 4. Losing NaN categories

```python
import numpy as np

df.loc[0, 'passenger_count'] = np.nan
df.loc[1, 'passenger_count'] = np.nan
df.loc[2, 'passenger_count'] = np.nan
df.loc[3, 'passenger_count'] = np.nan
df.loc[4, 'passenger_count'] = np.nan
```

```python
df.head()
```

```python
df.groupby('passenger_count')['trip_distance'].mean()
```

```python
# if we want to include NaN values in our categorical values, then we need to say dropna=False

df.groupby('passenger_count', dropna=False)['trip_distance'].mean()
```

# 5. Multiple column grouping



```python
# a good rule of thumb in Pandas is: If you can specify one column (as a string), then you can specify
# many columns (as a list of strings)

df.groupby('passenger_count')[['total_amount', 'trip_distance']].mean()
```

```python
df.groupby(['VendorID', 'passenger_count'])[['total_amount', 'trip_distance']].mean()
```

```python
df.groupby(['VendorID', 'passenger_count'])[['total_amount', 'trip_distance']].agg(['mean', 'std'])
```
